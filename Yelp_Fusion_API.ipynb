# Business Match

Similarly, you can match businesses by giving the name of a business and its address as you can see in this example

```
# Searching for Punjab Palace by giving name and Address 
url='https://api.yelp.com/v3/businesses/matches'
params={'name':"Punjab Palace","address1": "109 Brighton Ave",
     "address2": "",
     "address3": "",
     "city": "Allston",
     "state": "MA",
     "zip_code": "02134",
     "country": "US"}
req = requests.get(url, headers=headers, params=params)
print('The status code is {}'.format(req.status_code))
print(req.text)
```

Extracting information 

```
#Searching by term and Location
url='https://api.yelp.com/v3/businesses/search'
params={'term':'bookstore', 'location':'New York City'}
```

```
req = requests.get(url, params=params, headers=headers)
```

```
parsed = json.loads(req.text)
```

```
print(json.dumps(parsed, indent=4))
```

So you have to extract businesses first into a Python list and then loop over the businesses in the list

```
businesses = parsed["businesses"]
 
for business in businesses:
    print("Name:", business["name"])
    print("Rating:", business["rating"])
    print("Address:", " ".join(business["location"]["display_address"]))
    print("Phone:", business["phone"])
    print("\n")
```

Extract Business Reviews

```
id = business["id"]
```

```
url = "https://api.yelp.com/v3/businesses/" + id + "/reviews"
req = requests.get(url, headers=headers)
parsed = json.loads(req.text)
```

```
print(json.dumps(parsed, indent=4))
```

```
reviews = parsed["reviews"]
```

```
for review in reviews:
    print("User:", review["user"]["name"], "Rating:", review["rating"], "Review:", review["text"], "\n")
 
```
